```python
fig = plt.figure(figsize=(8, 8), dpi=160)
G1 = nx.DiGraph() #or G = nx.MultiDiGraph()
G1.add_nodes_from(range(1, 16))
G1.add_weighted_edges_from(acc)
ax1 = fig.add_subplot(1, 1, 1)
ax1.axis ('off') 
pos = nx.circular_layout(G1)
nx.draw_networkx(G1, pos, with_labels = True)
edge_labels=dict([((u,v,),d['weight'])
             for u,v,d in G1.edges(data=True)])
edge_labels=dict([((u,v,),d['weight'])
             for u,v,d in G1.edges(data=True)])
nx.draw_networkx_edge_labels(G1, pos, edge_labels=edge_labels, label_pos=0.3, font_size=5)
plt.savefig('dsselection.pdf')
```
